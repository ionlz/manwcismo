学习笔记：PMPP 书籍第 1-3 章的讲座总结
1. 简介

    动机：利用 GPU 提高数据处理应用的计算能力。GPU 被广泛应用于模拟和处理各种数据，包括股票市场、自驾车、游戏、天气预测、蛋白质折叠和机器人等领域。
    历史背景：传统软件开发以顺序执行为主，但随着 CPU 时钟频率和内存速度的提升遇到瓶颈，转向并行计算成为必然。多核 CPU 的引入要求开发者掌握多线程编程。
    并行计算的挑战：并行算法设计比顺序算法更复杂，需考虑数据依赖性和同步问题。性能往往受限于内存带宽，而非处理器速度。

2. 异构数据并行计算

    异构计算：指在 CPU 和 GPU 上协同处理计算任务。数据并行性是核心原则，即将计算任务分解为可独立执行的计算单元。
    CUDA 编程模型：
        基本概念：CPU 称为主机，GPU 称为设备。CUDA C 语言扩展了 C 语言，允许在 GPU 上编写并行代码。
        线程网格：CUDA 中线程按层次结构组织，分为网格和块，每个块中包含多个线程。线程通过内置变量（如 blockIdx 和 threadIdx）标识其在网格中的位置。
        内存管理：包括全局内存、共享内存等多种内存类型。CUDA 提供了内存分配和释放函数（如 cudaMalloc 和 cudaFree），以及内存复制函数（如 cudaMemcpy）。

3. 多维网格与数据

    多维数据处理：CUDA 支持多维网格和线程块，方便处理多维数据（如图像）。每个线程通常处理一个输出单元。
    内存布局：多维数组在内存中的存储是线性的，常见的布局有行主序和列主序。不同布局影响内存访问效率。

实例分析

    向量加法：利用 GPU 并行计算实现向量加法，每个线程负责计算一个输出单元。
    图像灰度化：将 RGB 图像转换为灰度图像，每个线程处理一个像素值的转换。
    图像模糊：实现均值滤波，每个线程计算一个像素周围的平均值，处理边界条件。

编程要点

    CUDA 内核函数：必须使用 __global__ 声明，表示在 GPU 上执行。内核函数的调用参数包括网格和块的大小。
    线程同步：并行程序中需要考虑线程间的同步和数据一致性问题。

练习和思考

    索引计算：理解如何利用 blockIdx 和 threadIdx 计算数据索引。
    性能优化：考虑内存访问模式和线程调度对性能的影响。

总结

    并行计算基础：理解并行计算和 CUDA 编程模型是进行 GPU 编程的基础。
    实践应用：通过实际编程练习，掌握如何将顺序算法转化为并行算法。

这份笔记总结了 GPU 并行计算的基本概念、CUDA 编程模型以及多维数据处理方法，并结合实例解释了如何应用这些知识。
